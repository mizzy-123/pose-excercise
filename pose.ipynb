```python
import cv2
import mediapipe as mp
import pandas as pd
import numpy as np

# Inisialisasi MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()
mp_drawing = mp.solutions.drawing_utils

# File input video (ganti dengan video yang ingin digunakan)
video_path = "./video/pushup3.mp4"  # Ubah ini dengan path video kamu
cap = cv2.VideoCapture(video_path)

# File untuk menyimpan data
csv_file = "pushup_data3.csv"
columns = [f"x_{i}" for i in range(33)] + [f"y_{i}" for i in range(33)] + [f"z_{i}" for i in range(33)] + ["label"]

# Inisialisasi dataframe
try:
    df = pd.read_csv(csv_file)
except FileNotFoundError:
    df = pd.DataFrame(columns=columns)

label = "pushup_up"  # Ubah sesuai kebutuhan
label2 = "pushup_down"

speed = 1

print("Tekan 's' untuk menyimpan data, 'q' untuk keluar")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Video selesai atau tidak bisa dibuka.")
        break
    
    # Konversi frame ke RGB
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(image)

    # Gambar landmark di video
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    if results.pose_landmarks:
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        key = cv2.waitKey(0) & 0xFF

        # Simpan data jika ditekan 's'
        if key == ord('s'):
            print(f"Continue")
        
        if key == ord('d'):
            landmarks = results.pose_landmarks.landmark
            data = [landmark.x for landmark in landmarks] + \
                   [landmark.y for landmark in landmarks] + \
                   [landmark.z for landmark in landmarks] + [label]

            df.loc[len(df)] = data
            df.to_csv(csv_file, index=False)
            print(f"Data dari frame ini disimpan dengan label: {label}")
        
        if key == ord('f'):
            landmarks = results.pose_landmarks.landmark
            data = [landmark.x for landmark in landmarks] + \
                   [landmark.y for landmark in landmarks] + \
                   [landmark.z for landmark in landmarks] + [label2]

            df.loc[len(df)] = data
            df.to_csv(csv_file, index=False)
            print(f"Data dari frame ini disimpan dengan label: {label2}")

    
    # Tampilkan video
    cv2.imshow('Push-Up Data Collection (Video)', image)

    
    # Tekan 'q' untuk keluar
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print("Selesai! Data push-up telah disimpan.")
```
